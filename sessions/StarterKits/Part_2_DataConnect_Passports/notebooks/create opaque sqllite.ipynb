Replace the named drs uris with numbered drs uris.

```python
import sqlite3
db_name = "../ga4gh-starter-kit.dev.db"
connection = sqlite3.connect(db_name)
```

```python
import json

with open("../drs_map.json") as f:
    drs_map = json.load(f)
```

```python
def fix_row(line):
    key_list = ['cram_drs_uri', 'crai_drs_uri', 'bundle_drs_uri']
    sample = json.loads(line)
    for k in key_list:
        # get the part after the last /
        drs_id = sample[k].split('/')[-1]
        sample[k] =  f"drs://localhost:5000/{drs_map[drs_id]}"
    #print(json.dumps(sample, indent=3))
    return sample
```

```python
cursor = connection.cursor()
rows = cursor.execute(
    "SELECT id, json_data FROM one_thousand_genomes_sample WHERE id > 5",

).fetchall()
id = str(rows[0][0])
print(rows[0][1])
```

```python
for r in rows:
    id =str(r[0])
    new_json = fix_row(r[1])
    #print(new_json)
    jstring = json.dumps(new_json)
    #print(jstring)
    cursor.execute(
    "UPDATE one_thousand_genomes_sample SET json_data = ? WHERE id = ?",
    (jstring,id))
    connection.commit()
    print(id)



```
